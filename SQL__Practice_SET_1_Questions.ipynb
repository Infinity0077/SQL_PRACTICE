## 6. Aggregate Functions & ROUND

27. Show the average age of all students (rounded to 2 decimals).

```python
# 27. Show the average age of all students (rounded to 2 decimals).
%%sql
SELECT ROUND(AVG(age), 2) AS avg_age
FROM student;
```

```python
# 28. Show the total number of international students.
%%sql
SELECT COUNT(inter_dom) AS total_inter_students
FROM student
WHERE inter_dom = "Inter";
```

```python
# 29. Show the minimum and maximum PHQ-9 (todep) score.
%%sql
SELECT MIN(todep) AS minimum_todep_score, MAX(todep) AS maximum_todep_score
FROM student;
```

```python
# 30. Show the average, sum, and count of 'tosc' for all students.
%%sql
SELECT AVG(tosc) AS avg_tosc, SUM(tosc) AS sum_of_tosc, COUNT(tosc) AS count_of_tosc
FROM student;
```

```python
# 31. Show the average length of stay for domestic students (rounded to 1 decimal).
%%sql
SELECT ROUND(AVG(stay), 1) AS avg_stay_dom
FROM student
WHERE inter_dom = "Dom";
```

```python
# 33. Show the average todep for each region (rounded to 2 decimals).
%%sql
SELECT region, ROUND(AVG(todep),2) AS avg_todep_per_region
FROM student
GROUP BY region;
```

## 8. Grouping (GROUP BY, HAVING)

37. Show the number of students in each region.
38. Show the average todep for each academic level.
39. Show the maximum length of stay for each gender.
40. Show the number of students in each Japanese proficiency category (japanese_cate).
41. Show the number of students grouped by region and gender.
42. Show all regions where there are more than 5 students.
43. Show the average toas (acculturative stress) for each region where the average is above 20.
44. Show all academic levels with more than 10 students.

```python
# 37. Show the number of students in each region.
%%sql
SELECT region, COUNT(id) AS students_per_region
FROM student
GROUP BY region;
```

```python
# 38. Show the average todep for each academic level.
%%sql
SELECT academic , AVG(todep) AS avg_todep_per_academic
FROM student
GROUP BY academic;
```

```python
# 39. Show the maximum length of stay for each gender.
%%sql
SELECT gender, MAX(stay) AS max_stay_per_gender
FROM student
GROUP BY gender;
```

```python
# 40. Show the number of students in each Japanese proficiency category (japanese_cate).
%%sql
SELECT japanese_cate, COUNT(id) AS total_students_per_proficiency
FROM student
GROUP BY japanese_cate;
```

```python
# 41. Show the number of students grouped by region and gender.
%%sql
SELECT region, gender, COUNT(id) AS no_of_student_per_region_and_gender
FROM student
GROUP BY region,gender;
```

```python
# 42. Show all regions where there are more than 5 students.
%%sql
SELECT region, COUNT(id) FROM student
GROUP BY region
HAVING COUNT(id) >5;
```

```python
# 43. Show the average toas (acculturative stress) for each region where the average is above 20.
%%sql
SELECT region, AVG(toas) AS avg_toas
FROM student
GROUP BY region
HAVING AVG(toas) >20;
```

```python
# 44. Show all academic levels with more than 10 students.
%%sql
SELECT academic, COUNT(id)
FROM student
GROUP BY academic
HAVING COUNT(id) > 10;
```

```python
# 55. Show the number of female graduate students in each region who have a depression score above 10.
%%sql
SELECT region, COUNT(*)
FROM student
WHERE todep > 10 AND gender ="Female" AND academic = "Grad"
GROUP BY region;
```

```python
# 56. Show the average todep for each region and academic combination, ordered by average todep descending.
%%sql
SELECT region, academic, AVG(todep) AS avg_todep
FROM student
GROUP BY region, academic
ORDER BY avg_todep DESC;
```

```python
# 57. Show all students whose todep is above the average todep for all students (use a subquery).
%%sql
SELECT * FROM student
WHERE todep > (
      SELECT AVG(todep)
      FROM student
);
```

```python
# 58. Show all students whose age is exactly twice their length of stay.
%%sql
SELECT * FROM student
WHERE age  = 2 * stay;
```

```python
# 59. Show the percentage of students in each region (as a decimal, not percentage).
%%sql
SELECT region, 1.0 * COUNT(*) / (SELECT COUNT(*) FROM student) AS region_fraction
FROM student
GROUP BY region;
```

```python
# 60. Show the average toas for students who have 'High' proficiency in both Japanese and English.
%%sql
SELECT AVG(toas) AS avg_toas
FROM student
WHERE japanese_cate = "High" AND english_cate = "High";
```
